```python
import requests
import json
import pandas as pd

td_consumer_key= 'RI2UBDPFALITZGOC8PVR3FPC19NLYQF3'
base_url = 'https://api.tdameritrade.com/v1/marketdata/chains?&symbol={stock_ticker}'
endpoint = base_url.format(stock_ticker = 'AAPL')
page = requests.get(url=endpoint, 
            params={'apikey' : td_consumer_key})
content = json.loads(page.content)
```

```python
c=content['callExpDateMap']
dfc=pd.DataFrame.from_dict(c, orient='index').stack().to_frame()
dfc=pd.DataFrame(dfc[0].values.tolist(), index=dfc.index)
dfc=pd.DataFrame(dfc[0].values.tolist(), index=dfc.index)
```

```python
p=content['putExpDateMap']
dfp=pd.DataFrame.from_dict(p, orient='index').stack().to_frame()
dfp=pd.DataFrame(dfp[0].values.tolist(), index=dfp.index)
dfp=pd.DataFrame(dfp[0].values.tolist(), index=dfp.index)
```

```python
df=pd.concat([dfp,dfc])
```
